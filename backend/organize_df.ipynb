```
import pandas as pd
import json

with open('/content/candidates.json') as f:
   data = json.load(f)
```

```
candidatos = data['candidates']
jobs = data['jobs']
```

```
techs = []
for candidato in candidatos:
  for tech in candidato['technologies']:
    techs.append(tech['name'])
```

```
columns = ['id', 'city', 'experience']
unique_techs = list(dict.fromkeys(techs))
columns.extend(unique_techs)
```

```
data = np.zeros((len(candidatos),154))
df = pd.DataFrame(data, columns = columns)
for i, candidate in enumerate(candidatos):
  df['id'][i] = str(candidate['id'])
  df['city'][i] = candidate['city']
  df['experience'][i] = candidate['experience']
  for candidato in candidate['technologies']:
    if candidato['name'] in columns:
      df[candidato['name']][i] = 1
    else:
      df[candidato['name']][i] = 0
    
```
